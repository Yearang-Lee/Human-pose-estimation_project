# 3. 준비

```python
def makeSureBackGroundClearConfirm():

    ######## Options #########
    fontColor = (0, 0, 255); fontScale = 1;  lineType = 2; font = cv2.FONT_HERSHEY_SIMPLEX
    textPositionOriginal = (600, 100)
    welcomeText = 'Ready and Press Enter!!!!'   
    betweenLineSpace = 100

    welcomeTextArray = welcomeText.split('/n')
    
    videoCam = cv2.VideoCapture(0)     # index : 카메라 장치 번호 (0 부터 차례로 증가)
    videoCam.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
    videoCam.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)
    
    capture = cv2.VideoCapture("cut_video.mp4")
    
    # Check web cam is working
    if videoCam.isOpened():       # 캡쳐 객체 연결 확인
        # ret : 프레임 읽기 성공 또는 실패 여부, True / False
        # originalVid : 프레임 이미지, Numpy 배열 또는 None
        ret, originalVid = videoCam.read()     # 다음 프레임 읽기
    else:
        ret = False
        
    # Original Video
    while ret:  # 프레임 읽기가 성공 했다면~
        ret, originalVid = videoCam.read()

        if(capture.get(cv2.CAP_PROP_POS_FRAMES) == capture.get(cv2.CAP_PROP_FRAME_COUNT)):
            capture.open("cut_video.mp4")
            
        _, frame = capture.read()
        frame = cv2.resize(frame, (500, 720))
        originalVid[0:720, 0:500] = frame
        
        
        press_img = cv2.imread('img/press.png')
        press_img = cv2.resize(press_img, (160, 720))
        originalVid[0:720, 1120:1280] = press_img
        
        
        
        #### Write Text ####
        textPosition = textPositionOriginal
        for i in range(len(welcomeTextArray)):
            cv2.putText(originalVid, welcomeTextArray[i], textPosition, font, fontScale, fontColor, lineType)
            textPositionList = list(textPosition)    
            textPositionList[1] += betweenLineSpace;
            textPosition = tuple(textPositionList)
        cv2.imshow("Original", originalVid)
        inputKey = cv2.waitKey(1)  # 1ms 지연(40fps로 가정) 
        if inputKey == 13:
            global openGame; openGame=4
            break;
        elif inputKey == 27:
            break;
    cv2.destroyAllWindows()
    videoCam.release()
```

# Main

# 4. 동작 매칭

```python
def crop(image, w, f):
    return image[:, int(w * f): int(w * (1 - f))]

def main():
    #### load train model ####
    model = load_model('vgg19.h5')
    #img_dim = 224
    class_labels = ['1', '2', '3', '4', '5', '6', '7'] 
    
    #### load openpose(skeleton) model ####
    posemodel = get_testing_model()
    posemodel.load_weights('model/keras/model.h5')
    params, model_params = config_reader()
    
    black = cv2.imread('img/black.png')
    black = cv2.resize(black, (395, 700))
    
    #### option #### 
    factor = 0.1
    scale_search = [0.22, 0.25, .5, 1, 1.5, 2]  
    process_speed = 1
    scale_search = scale_search[0:process_speed]
    params['scale_search'] = scale_search
    resize_fac = 5
    
    cap = cv2.VideoCapture(0)
    cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)
    
    number = [ 1 , 2 , 3 , 4 , 5 , 6 , 7 , 8 , 9 ]   
    name_list = np.array(sorted(glob.glob("show_img/*")))
    i = 0
    
    while True:
        if i < len(number):
            name = name_list[number[i]-1]
            icon = cv2.imread('{}'.format(name))
            icon = cv2.resize(icon, (500, 720))
            order = number[i]
            ret, frame = cap.read()
            #print(frame.shape)
            width = frame.shape[1]
            height = frame.shape[0]
            #################### webcam space #################### 
            cv2.rectangle(frame, (328, 50), (645, 670), (255, 255, 255), 2)  # 100, 620
            
            
            #################### webcam openpose #################### 
            frame = crop(frame, width, factor)
            input_image = cv2.resize(frame, (0, 0), fx=1/resize_fac, fy=1/resize_fac, interpolation=cv2.INTER_CUBIC)
            input_image = cv2.cvtColor(input_image, cv2.COLOR_RGB2BGR)
            body_parts, all_peaks, subset, candidate = extract_parts(input_image, params, posemodel, model_params)
            frame = draw(frame, all_peaks, subset, candidate, resize_fac=resize_fac)
            black_frame = draw(black, all_peaks, subset, candidate, resize_fac=resize_fac)
            #################### frame effect #################### 
            frame = cv2.flip(frame, 1)
            black_frame = cv2.flip(black_frame, 1)
            frame = cv2.resize(frame, (1280, 720))
            frame[0:720, 0:500] = icon 
            
            #################### webcam 포즈 예측 #################### 
            img = cv2.cvtColor(black_frame, cv2.COLOR_BGR2RGB)
            img = cv2.resize(img, (700,395))     
            #cv2.imwrite('pred_img.jpg', img)  # 저장
            
            img = image.img_to_array(img)   
            img = np.expand_dims(img, axis=0)            
            img = img.astype('float32')/255
            pred = np.argmax(model.predict(img)) 
            #cv2.putText(frame, class_labels[pred], (40,600), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0,0,255), 2) 
            cv2.imshow('Webcam', frame)
            
            if str(class_labels[pred]) == str(order):  
                img = cv2.imread('img/success.png')
                cv2.imshow('img/success',img)
                cv2.waitKey(1000)
                cv2.destroyAllWindows()
                
                i += 1
                
                
            else:
                if cv2.waitKey(1) & 0xFF == ord('q'):
                    break
                ret, frame = cap.read()    
        else:
            break
                   
    cap.release()
    cv2.destroyAllWindows()
```

```python
openGame = 0;   

startGameInterface()

if(openGame==0):
    idolchoice()
    
if(openGame==1):
    singchoice()

if(openGame==2):
    memberchoice()

if(openGame==3):
    makeSureBackGroundClearConfirm()

if(openGame==4):
    if __name__=="__main__":
        main()
```
